# Central Dogma Sequence Analysis Demo

This notebook demonstrates DNA validation, transcription, translation, GC content analysis, and simple visualization.

```python
from pathlib import Path
import sys

project_root = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(project_root))

from central_dogma import validate_dna, transcribe_dna_to_rna, translate_rna_to_protein
from analysis import calculate_gc_content, count_bases, count_codons
from visualization import plot_base_composition, plot_codon_frequency, plot_amino_acid_composition
```

```python
dna = "ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG"
dna
```

```python
is_valid = validate_dna(dna)
is_valid
```

```python
rna = transcribe_dna_to_rna(dna)
protein = translate_rna_to_protein(rna)
rna, protein
```

```python
gc_content = calculate_gc_content(dna)
base_count = count_bases(dna)
codon_count = count_codons(rna)
gc_content, base_count
```

```python
plot_base_composition(base_count)
```

```python
plot_codon_frequency(codon_count)
```

```python
plot_amino_acid_composition(protein)
```

## Short Interpretation

The DNA sequence is valid, can be transcribed into RNA, and translates into a protein sequence until the first stop codon. GC content and composition plots provide a simple summary of sequence characteristics.
